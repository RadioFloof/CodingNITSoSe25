```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Load the Titanic dataset
df = pd.read_csv("C:/Users/rforner/CodingNITSoSe25/titanic/train.csv")

# Drop rows with missing Age or Sex or Survived
df = df.dropna(subset=["Age", "Sex", "Survived"])

# Encode Sex column as binary (0 = male, 1 = female)
df['Sex'] = df['Sex'].map({"male": 0, "female": 1})

# Train a logistic regression model to predict survival based on Age and Sex
X = df[["Age", "Sex"]]
y = df["Survived"]

# Fit the model
model = LogisticRegression()
model.fit(X, y)

# Predict probabilities of survival
df['SurvivalProb'] = model.predict_proba(X)[:, 1]

# Plot the probability of survival by Sex and Age
plt.figure(figsize=(10, 6))

# Plot the probabilities for each Sex
sns.lineplot(x="Age", y="SurvivalProb", data=df[df['Sex'] == 0], label="Male", color="blue")
sns.lineplot(x="Age", y="SurvivalProb", data=df[df['Sex'] == 1], label="Female", color="pink")

# Customize the plot
plt.title("Probability of Survival by Age and Sex")
plt.xlabel("Age")
plt.ylabel("Probability of Survival")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_titanic_routes_with_counts(csv_path):
    """
    Reads Titanic passenger data from csv_path, plots routes for each passenger,
    and counts the number of passengers for each unique start-to-end pair.
    Labels each line with the count of passengers for that route.
    """
    # 1. Load Titanic data
    df = pd.read_csv(csv_path)
    df = df.dropna(subset=['Embarked', 'Sex', 'Survived'])  # Keep necessary columns

    # 2. Port coordinates (embarkation points)
    coords = {
        'Southampton': (-1.404351, 50.9097),  # Southampton
        'Cherbourg': (-1.6197, 49.6367),   # Cherbourg
        'Queenstown': (-8.2812, 51.8502)    # Queenstown (Cobh)
    }

    # 3. Destination coordinates
    titanic_site = (-49.95, 41.73)     # Titanic wreck site (lon, lat)
    new_york     = (-74.0060, 40.7128) # New York (lon, lat)

    # 4. Set up the plot
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.set_xlim(-80, 5)
    ax.set_ylim(40, 56)
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')
    ax.set_title('Titanic Passengers: Survivors → New York, Non-survivors → Titanic Site')

    # 5. Count passengers for each route pair (embarkation -> destination)
    route_counts = {}

    for _, p in df.iterrows():
        port = p['Embarked']
        if port not in coords:
            continue
        lon0, lat0 = coords[port]
        # Determine the destination based on survival
        lon1, lat1 = new_york if p['Survived'] == 1 else titanic_site
        route = (port, 'Survived' if p['Survived'] == 1 else 'Non-survived')

        # Update count for this route pair
        if route not in route_counts:
            route_counts[route] = 0
        route_counts[route] += 1

        # Color and marker for the line
        color = 'red' if p['Sex'] == 'female' else 'blue'
        marker = 'o' if p['Survived'] == 1 else '*'

        # Plot the line for the passenger
        ax.plot([lon0, lon1], [lat0, lat1], color=color, lw=0.5)
        ax.plot(lon1, lat1, marker=marker, ms=5, color=color, alpha=0.6)

    # 6. Mark embarkation ports on the plot
    for name, (lon, lat) in coords.items():
        ax.plot(lon, lat, marker='s', ms=8, color='black')
        ax.text(lon + 1, lat + 0.5, name, fontsize=10)

    # 7. Mark Titanic site and New York destinations
    ax.plot(*titanic_site, marker='X', ms=12, color='black')
    ax.text(titanic_site[0] + 1, titanic_site[1], 'Titanic Site', fontweight='bold')

    ax.plot(*new_york, marker='D', ms=8, color='black')
    ax.text(new_york[0] + 1, new_york[1], 'New York', fontweight='bold')

    # 8. Label each route with the count of passengers
    for (port, survival), count in route_counts.items():
        if survival == 'Survived':
            lon1, lat1 = new_york
        else:
            lon1, lat1 = titanic_site
        # Calculate midpoint for labeling
        lon0, lat0 = coords[port]
        mid_lon = (lon0 + lon1) / 2
        mid_lat = (lat0 + lat1) / 2
        ax.text(mid_lon, mid_lat, str(count), fontsize=10, color='black', weight='bold')

    # 9. Display the plot
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

if __name__ == '__main__':
    # Path to Titanic train CSV file
    csv_file = 'titanic/train.csv'  # Make sure this is the correct file path
    plot_titanic_routes_with_counts(csv_file)
```
