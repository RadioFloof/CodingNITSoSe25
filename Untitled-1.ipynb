```python
import math

# Step 1: Create a list from 1 to 100
my_list = list(range(1, 101))

# Step 2: Create a list of (sqrt, sqrt+1)
sqrt_plus_one_list = []
for x in my_list:
    sqrt_x = math.sqrt(x)
    sqrt_plus_one = sqrt_x + 1
    sqrt_plus_one_list.append((sqrt_x, sqrt_plus_one))

# Step 3: Divide each number by sqrt and sqrt+1 using modulo
modulo_results = []
for idx, x in enumerate(my_list):
    sqrt_x, sqrt_plus_one = sqrt_plus_one_list[idx]
    mod_by_sqrt = x % sqrt_x
    mod_by_sqrt_plus_one = x % sqrt_plus_one
    modulo_results.append((mod_by_sqrt, mod_by_sqrt_plus_one))

# Step 4: Create a new list where modulo with sqrt == 0
filtered_list = []
for idx, x in enumerate(my_list):
    mod_by_sqrt, _ = modulo_results[idx]
    if mod_by_sqrt == 0:
        filtered_list.append(x)

# Step 5: Print everything nicely
print("Original list (1 to 100):")
print(my_list)
print("\nSquare roots and Square roots + 1:")
for pair in sqrt_plus_one_list:
    print(pair)
print("\nModulo results (number % sqrt, number % (sqrt+1)):")
for mods in modulo_results:
    print(mods)
print("\nFiltered list (where number % sqrt == 0):")
print(filtered_list)
```

```python
import math

# Step 1: Create a list from 1 to 100
my_list = list(range(1, 101))

# Step 2: Define the function to check for prime numbers
def is_prime(n):
    
    
    
    # Checking divisibility using modulo with sqrt and sqrt+1
    sqrt_n = math.sqrt(n)
    for i in range(2, int(sqrt_n) + 1):
        if n % i == 0:
            return False
    return True

# Step 3: Apply the primality test to the list
primes = [x for x in my_list if is_prime(x)]

# Step 4: Print the list of prime numbers
print("Prime numbers from 1 to 100:")
print(primes)
```

```python
import csv

with open('advent_of_code_exe_1.txt', 'r') as txtfile, open('AOCE1.csv', 'w', newline='') as csvfile:
 
    writer = csv.writer(csvfile)
 
    for line in txtfile:
 
        parts = line.strip().split()  # split by space
 
        writer.writerow(parts)

 
# Open a file for reading

with open("advent_of_code_exe_1.txt", "r") as f:

    lines = f.readlines()

    print(lines)
 
# Define left_list and right_list based on the content of the file

left_list = [int(line.split()[0]) for line in lines]

right_list = [int(line.split()[1]) for line in lines]
 
# Sort both lists in ascending order

left_list.sort()

right_list.sort()
 
# Calculate the total distance by summing absolute differences

total_distance = 0

for left, right in zip(left_list, right_list):

    total_distance += abs(left - right)
 
# Print the total distance

print(f"The total distance between the left and right lists is: {total_distance}")
 
 
```

```python
def calculator():
    try:
        num1 = float(input("Enter first number: "))
        operator = input("Enter operator (+, -, *, /): ")
        num2 = float(input("Enter second number: "))

        if operator == '+':
            result = num1 + num2
        elif operator == '-':
            result = num1 - num2
        elif operator == '*':
            result = num1 * num2
        elif operator == '/':
            result = num1 / num2
        else:
            print("Invalid operator!")
            return

        print(f"Result: {result}")

    except ValueError:
        print("Invalid input! Please enter numbers only.")
    except ZeroDivisionError:
        print("Cannot divide by zero.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Run the calculator
calculator()
```

```python
def calculator():
    while True:
        try:
            num1_input = input("Enter first number (or type 'exit' to quit): ")
            if num1_input.lower() == 'exit':
                print("Goodbye!")
                break
            num1 = float(num1_input)

            operator = input("Enter operator (+, -, *, /): ")
            if operator not in ('+', '-', '*', '/'):
                print("Invalid operator!")
                continue

            num2_input = input("Enter second number: ")
            if num2_input.lower() == 'exit':
                print("Goodbye!")
                break
            num2 = float(num2_input)

            if operator == '+':
                result = num1 + num2
            elif operator == '-':
                result = num1 - num2
            elif operator == '*':
                result = num1 * num2
            elif operator == '/':
                result = num1 / num2

            print(f"Result: {result}")

        except ValueError:
            print("Invalid input! Please enter numbers only.")
        except ZeroDivisionError:
            print("Cannot divide by zero.")
        except Exception as e:
            print(f"An unexpected error occurred: {e}")

# Run the calculator
calculator()
```

```python
def get_number(prompt):
    while True:
        try:
            value = input(prompt)
            if value.lower() == 'exit':
                print("Goodbye!")
                exit()
            return float(value)
        except ValueError:
            print("Invalid number! Please try again.")

def get_operator():
    while True:
        operator = input("Enter operator (+, -, *, /): ")
        if operator in ('+', '-', '*', '/'):
            return operator
        else:
            print("Invalid operator! Please enter +, -, *, or /.")

def calculator():
    while True:
        num1 = get_number("Enter first number (or type 'exit' to quit): ")
        operator = get_operator()
        num2 = get_number("Enter second number: ")

        try:
            if operator == '+':
                result = num1 + num2
            elif operator == '-':
                result = num1 - num2
            elif operator == '*':
                result = num1 * num2
            elif operator == '/':
                result = num1 / num2

            print(f"Result: {result}")
        except ZeroDivisionError:
            print("Cannot divide by zero. Please re-enter the second number.")
            num2 = get_number("Enter second number: ")
            try:
                result = num1 / num2
                print(f"Result: {result}")
            except ZeroDivisionError:
                print("Still division by zero. Operation cancelled.")

# Run it
calculator()
```
